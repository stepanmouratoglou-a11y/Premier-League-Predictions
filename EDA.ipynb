## **This is also a small copy of the main dataset to create a sorted Diverging Bar Chart for the Goal Difference of each club**

```python
goal_differences=dataset[['Team','Goal_Difference']].sort_values(by='Goal_Difference',ascending=True)
```

```python
plt.figure(figsize=(10,8), dpi=80)
colors = ['red' if diff<0 else 'green' for diff in goal_differences['Goal_Difference']]
plt.barh(y=goal_differences['Team'], width=goal_differences['Goal_Difference'], color=colors, height=0.8)
plt.axvline(0, color='black', linewidth=1.5, linestyle='-')
plt.xlabel("Goal Difference")
plt.ylabel("Clubs")
plt.yticks(fontsize=14)
plt.title("Goal Difference per club",fontdict={
    'size':20
})
plt.grid(linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
team_colors = {
    'Man City':'cyan',
    'Arsenal':'firebrick',
    'Liverpool':'red',
    'Man United':'red',
    'Chelsea':'darkblue',
    'Aston Villa':'m',
    'Fulham':'gray',
    'Brentford':'darkred',
    'Brighton':'dodgerblue',
    'Burnley':'blueviolet',
    'Bournemouth':'maroon',
    'Crystal Palace':'lightskyblue',
    'Everton':'mediumblue',
    'Leeds':'yellow',
    'Newcastle':'black',
    'Sunderland':'lightcoral',
    'Tottenham':'white',
    'West Ham':'firebrick',
    'Wolves':'gold'
}
```

```python
plt.figure(figsize=(10,6),dpi=100)
for index,row in dataset.iterrows():
  team_name=row['Team']
  team_color=team_colors.get(team_name,'hotpink')
  plt.scatter(row['Shots_Per_Game'],row['Goals_Per_Game'],color=team_color,s=120,edgecolor='black',zorder=2)
  plt.annotate(
        team_name,
        (row['Shots_Per_Game'], row['Goals_Per_Game']),
        textcoords="offset points",
        xytext=(0, 12),
        ha='center',
        fontsize=8
    )
plt.xlabel("Shots Per Game",fontsize=12)
plt.ylabel("Goals Per Game",fontsize=12)
plt.title("Relation between shots per game and goals per game\n (Each point corresponds to a different club)\n\n",
          fontsize=15,
          fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.5, zorder=1)
plt.show()
```

## **Creating a copy of some feature of our dataset to visualize the most "aggressive" clubs**

```python
yellow_cards=dataset[['Team','Yellow_Cards_Per_Game','Yellow_Cards','Fouls_commited']]
yellow_cards=yellow_cards.sort_values(by='Yellow_Cards_Per_Game',ascending=True)
```

```python
fig , axes =plt.subplots(nrows=1, ncols=2,figsize=(15,5))
axes[0].barh(yellow_cards['Team'],yellow_cards['Yellow_Cards'],color='yellow')
axes[0].set_xlabel("Total Yellow Cards")
axes[0].set_label("Clubs")
axes[0].set_title("Total Yellow Cards Per Club")

axes[1].barh(yellow_cards['Team'],yellow_cards['Yellow_Cards_Per_Game'],color='yellow')
axes[1].set_xlabel("Yellow Cards per Game")
axes[1].set_ylabel("Club")
axes[1].set_title("Yellow cards Per Game Per Club")

plt.tight_layout()
plt.show()
```

#### After analyzing the "aggression" of the clubs we can see below, that the Relation between fouls commited and yellow card seems pretty nice,except the fact that a very suspicious club from Pireaus has commited a 345 fouls , and has approximately 20 yellow card , whilst another club with 350 fouls (almost the same amount) has 4 times more yellow cards.
**Very suspecious**

```python
plt.figure()
plt.barh(yellow_cards['Team'],yellow_cards['Fouls_commited'],color='purple')
plt.ylabel("Club")
plt.xlabel("Total Fouls Commited")
plt.title("Fouls Commited Per Club")
plt.show()
```

```python
plt.figure(figsize=(8,5))
for index,row in yellow_cards.iterrows():
  team_name=row['Team']
  team_color=team_colors.get(team_name,'hotpink')
  plt.scatter(row['Fouls_commited'],row['Yellow_Cards'],color=team_color,s=120,edgecolor='black',zorder=2)
  plt.annotate(
        team_name,
        (row['Fouls_commited'], row['Yellow_Cards']),
        textcoords="offset points",
        xytext=(0, 12),
        ha='center',
        fontsize=6
    )
plt.xlabel("Total Fouls Commited")
plt.ylabel("Total Yellow Cards Received")
plt.title("Relation between Fouls Commited and Yellow Cards")
plt.tight_layout()
plt.show()
```
